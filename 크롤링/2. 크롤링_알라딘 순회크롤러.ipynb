```python
# 크롤링 작업을 위한 라이브러리 임포트
from bs4 import BeautifulSoup # 웹에서 가져온 데이터를 정제하는 역할
import requests # 파이썬에서 다이렉트로 자료를 요처앟는 역할
from selenium import webdriver # 크롬창을 켜주는 역할
import time # 코드 중간에 딜레이를 넣는 역할
from selenium.webdriver.common.by import By
import codecs
```

```python
# 셀레니움 기본세팅을 해주세요.
# 알라딘 베스트셀러 페이지에 접속하도록 만들어주시면 됩니다,
browser= webdriver.Chrome('chromedriver.exe')


# 브라우저 켜기
# 알라딘 베스트셀러 페이지 접속
browser.get("https://www.aladin.co.kr/shop/common/wbest.aspx?BranchType=1")

# 소스코드 변수에 저장하기
source = browser.page_source
```

```python
# 소스코드 파싱
aladin_code = BeautifulSoup(source, "html.parser")
```

```python
# 순위별로 데이터를 가져오기 위해 처리
# 태그명이 div 이면서 클래스명이 ss_book_box 인 요소를 가져와야함
div_box_list = aladin_code.find_all("div", class_="ss_book_box")
len(div_box_list)
```

```python
# li_list의 몇 번째가 제목이고, 가격이고, 저자인지 알면 교보처럼 출력하는데 어려움이 없음
for div in div_box_list:
    li_list = div.find_all("li")
    # title = div.find_all("b")
    # 해당 li_list중에 우리가 찾는 제목도 있고 가격도 있고, 저자도 있음
    # for li in li_list:
    #    print(li.text)
    if(len(li_list) > 7):
        print('제목 : ' + li_list[1].text)
        print('저자 : ' + li_list[2].text)
        print('가격 : ' + li_list[3].text)
        print("----------------------------")
    else:
        print('제목 : ' + li_list[0].text)
        print('저자 : ' + li_list[1].text)
        print('가격 : ' + li_list[2].text)
        print("----------------------------")
```

```python
# browser 변수를 이용해서 알라딘 베스트셀러 페이지를 열어주시고
# 그 와중에 각 페이지 이동당 2초씩 딜레이를 주시면 됩니다.
# 반복문을 이용해 마지막 페이지까지 순회시켜주세요.
browser = webdriver.Chrome("chromedriver")
aladin_txt = codecs.open("c:/crawler2/aladin.txt", encoding="utf-8", mode="w")

# 책 순위를 찍습니다.
book_rank = 0

for i in range(1, 21):
    browser.get(
        "https://www.aladin.co.kr/shop/common/wbest.aspx?BestType=Bestseller&BranchType=1&page=%s" % i)
    time.sleep(2)
    
    # 소스코드 저장
    source = browser.page_source
    # 파싱
    aladin_code = BeautifulSoup(source, "html.parser")
    # 출력
    div_box_list = aladin_code.find_all("div", class_="ss_book_box")
    len(div_box_list)
    # li_list의 몇 번쨰가 제목이고, 가격이고, 저자인지 알면 교보처럼 출력하는데 어려움이 없음
    for div in div_box_list:
        # 책이 바뀔때마다 책 순위를 1위씩 더해줌
        book_rank += 1
        li_list = div.find_all("li")
        # 해당 li_list중에 우리가 찾는 제목도 있고, 가격도 있고, 저자도 있음
        if(len(li_list) > 7):
            aladin_txt.write("%s등 책" % book_rank + "\n")
            aladin_txt.write(li_list[1].text + "\n")
            aladin_txt.write(li_list[2].text + "\n")
            aladin_txt.write(li_list[3].text + "\n")
        else :
            aladin_txt.write("%s등 책" % book_rank + "\n")
            aladin_txt.write(li_list[0].text + "\n")
            aladin_txt.write(li_list[1].text + "\n")
            aladin_txt.write(li_list[2].text + "\n")
aladin_txt.close()
```
